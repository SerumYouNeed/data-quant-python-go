```python
# Import bibliotek
import numpy as np
import matplotlib.pyplot as plt

# Wektor cen akcji (5 dni)
prices = np.array([100, 102, 105, 103, 107])
print("Ceny:", prices)
print("Typ:", prices.dtype)
print("Kształt:", prices.shape)

# Dzienne zmiany: różnica między kolejnymi dniami
returns = prices[1:] / prices[:-1] - 1
print("Dzienne zmiany procentowe:", returns)

# Średni zwrot i zmienność
mean_return = np.mean(returns)
volatility = np.std(returns)

print("Średni zwrot:", mean_return)
print("Zmienność (STD):", volatility)

# Porównanie dwóch portfeli
portfolio_1 = np.array([0.2, 0.3, 0.5])  # udziały w akcjach
returns_1 = np.array([0.01, -0.005, 0.02])

portfolio_2 = np.array([0.4, 0.4, 0.2])
returns_2 = np.array([0.015, -0.01, 0.025])

# Łączny zwrot portfela
total_1 = np.dot(portfolio_1, returns_1)
total_2 = np.dot(portfolio_2, returns_2)

print("Zwrot portfela 1:", total_1)
print("Zwrot portfela 2:", total_2)

# Symulacja 100 dniowych zwrotów z rozkładu normalnego
simulated_returns = np.random.normal(loc=0.001, scale=0.01, size=100)

# Skumulowany zwrot portfela (wartość inwestycji w czasie)
cumulative = np.cumprod(1 + simulated_returns)

plt.plot(cumulative)
plt.title("Symulowany wzrost inwestycji")
plt.xlabel("Dzień")
plt.ylabel("Kapitał")
plt.grid(True)
plt.show()
```
